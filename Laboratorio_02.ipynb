# Actividad 1
## Distribución uniforme
Se requiere de la librerías **pandas**, **matplotlib.pyplot**, **numpy** y de **scipy.stats** la función **uniform** para producir las probabilidades y la figura.
1. Se crea el *data frame* con los valores de la VA del ejemplo y sus probabilidades asociadas. 
2. Se configura el gráfico con opciones básicas y se despliega.

```
df[df['x']<=2]
```

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import uniform 
 
t1, t2 = 1, 4
df = pd.DataFrame({'x': np.linspace(t1-1, t2+1, 1000)})
df['p'] = uniform.pdf(df.x, t1, t2-t1) # Origen (t1) y longitud (t2-t1)
df['F'] = uniform.cdf(df.x, t1, t2-t1)
print("Media: ", round(uniform.mean(t1, t2-t1), 2))
print("Varianza: ", round(uniform.var(t1, t2-t1), 2))
print("Desviación estándar: ", round(uniform.std(t1, t2-t1),2), "\n")
# pdf
plt.figure(figsize=(12,6))
plt.plot(df.x, df.p)
plt.xlim([0, 4])
plt.ylim([0, 0.7])
plt.ylabel('f(x)', fontsize=20)
plt.xlabel('x', fontsize=20)
plt.show()
# cdf
plt.figure(figsize=(12,6))
plt.plot(df.x, df.F)
plt.xlim([0, 4])
plt.ylim([0, 1.1])
plt.ylabel('F(x)', fontsize=20)
plt.xlabel('x', fontsize=20)
plt.show()
```

# Actividad 2:
## Distribución exponencial
Ahora, además de las anteriores, de **scipy.stats** se requiere de la función **expon**. Nótese que para este caso se debe utilizar el parámetro como $1/\lambda$ en lugar de $\lambda$. Ingresa a la función como el parámetro *scale*.

```
from scipy.stats import expon 
l = 2
df = pd.DataFrame({'x': np.linspace(0, 5, 250)})
df['p'] = expon.pdf(df.x, scale=1/l)
df['F'] = expon.cdf(df.x, scale=1/l)
print("Media: ", round(expon.mean(scale=1/l),2))
print("Varianza: ", round(expon.var(scale=1/l), 2))
print("Desviación estándar: ", round(expon.std(scale=1/l),2), "\n")
# pdf
plt.figure(figsize=(12,6))
plt.plot(df.x, df.p)
plt.xlim([0, 5])
plt.ylim([0, 2.1])
plt.ylabel('f(x)', fontsize=20)
plt.xlabel('x', fontsize=20)
plt.show()
# cdf
plt.figure(figsize=(12,6))
plt.plot(df.x, df.F)
plt.xlim([0, 5])
plt.ylim([0, 1.1])
plt.ylabel('F(x)', fontsize=20)
plt.xlabel('x', fontsize=20)
plt.show()
```

# Actividad 3:
## Distribución normal
Ahora se requiere de **scipy.stats** la función **norm**. Esta es una actividad similar a las anteriores, pero un poco más compleja, ya que se construye un gráfico que superpone varias series de valores. 

```
from scipy.stats import norm 
 
m=4 # La media
s=2 # La desviación estándar
li = m-3*s
ls = m+3*s
df = pd.DataFrame({'x': np.linspace(li, ls, 1000)})
df['p'] = norm.pdf(df.x, m, s)
df['F'] = norm.cdf(df.x, m, s)
print(df)
print("Media: ", round(norm.mean(m, s),2))
print("Varianza: ", round(norm.var(m, s), 2))
print("Desviación estándar: ", round(norm.std(m, s),2), "\n")
# pdf
plt.figure(figsize=(12,6))
plt.plot(df.x, df.p)
plt.xlim([li, ls])
plt.ylim([0, 0.5])
plt.ylabel('f(x)', fontsize=20)
plt.xlabel('x', fontsize=20)
plt.show()
# cdf
plt.figure(figsize=(12,6))
plt.plot(df.x, df.F)
plt.xlim([li, ls])
plt.ylim([0, 1.1])
plt.ylabel('F(x)', fontsize=20)
plt.xlabel('x', fontsize=20)
plt.show()
```

```
print("P[X>=4] = ", round(norm.cdf(0.5), 4))
```

```
# La probabilidad de un intervalo cerrado o abierto es la misma en una VA continua
print("P[1<=X<=3] = ", round(2*norm.cdf(0.8) - 1, 4))
print("P[X>=4] = ", round(1-norm.cdf(1.6), 4))
print("P[X<=4] = ", round(norm.cdf(1.6), 4))
# Ahora buscando el valor de X tal que...
x = round(norm.ppf(1-0.05)*1.25+2, 4)
print("x:{P[X>x] = 0.05} = ", x)
print("Nótese que P[X>",x,"] = ", round(1-norm.cdf(x, 2, 1.25), 4))
# Otro
x = round(norm.ppf(0.875, 2, 1.25), 4)
print("x, xs:{P[xs<=X<=x] = 0.75} = ", x, x-2)
# Otro
x = round(norm.ppf(0.975, 2, 1.25), 4)
print("x, xs:{P[X<=xs o X>=x] = 0.05} = ", x, x-2)
```

Solución:

Los números aleatorios de una distribución normal estándar se pueden generar mediante la función 'np.random.normal' o mediante la función 'np.random.default_rng().normal'

```
import numpy as np
from scipy.stats import truncnorm
import matplotlib.pyplot as plt

#Forma 1
mu, sigma = 0, 1 # Media y desviación estándar
randomNums = np.random.normal(mu, sigma, 10000)
randomInts = np.round(randomNums)

axis = np.arange(start=min(randomInts), stop = max(randomInts) + 1)
plt.hist(randomInts, bins = axis)
plt.show()
```

```
#Forma 2
mu, sigma = 0, 1 # Media y desviación estándarn
s = np.random.default_rng().normal(mu, sigma, 10000)

count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```
