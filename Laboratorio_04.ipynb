### Tarea 4:
+ Si se tienen dos muestras de poblaciones distintas, construir un intervalo de confianza para la diferencia de medias.


```python
import numpy as np, statsmodels.stats.api as sms

X1 = np.arange(10,21)
X2 = np.arange(20,26.5,.5)

cm = sms.CompareMeans(sms.DescrStatsW(X1), sms.DescrStatsW(X2))
print('El intervalo de confianza para la diferencia de medias es: ', cm.tconfint_diff(usevar='unequal'))
```

+ Programe una ilustración del Teorema del límite central.

```python
import numpy 
import matplotlib.pyplot as plt 
  
num = [1, 10, 50, 100]   
means = []   
  
for j in num: 
    
    
    numpy.random.seed(1) 
    x = [numpy.mean( 
        numpy.random.randint( 
            -40, 40, j)) for _i in range(1000)] 
    means.append(x) 
k = 0
  
fig, ax = plt.subplots(2, 2, figsize =(8, 8)) 
for i in range(0, 2): 
    for j in range(0, 2): 
        
        ax[i, j].hist(means[k], 10, density = True) 
        ax[i, j].set_title(label = num[k]) 
        k = k + 1
```
